# 2. Pre-processing


```python
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)

    # Lemmatization (using ISRIStemmer as an approximation for Hebrew lemmatization)
    stemmer = ISRIStemmer()
    lemmatized = [stemmer.stem(token) for token in tokens]

    # Stop words removal
    hebrew_stop_words = set(stopwords.words('hebrew'))
    filtered = [word for word in lemmatized if word.lower() not in hebrew_stop_words]

    return filtered

processed_texts = [preprocess_text(text) for text in texts]
```

# 3. TF-IDF analysis and visualization

```python
def tfidf_analysis(texts):
    # Flatten the list of lists into a single list of strings
    flattened_texts = [' '.join(text) for text in texts]

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(flattened_texts)
    feature_names = vectorizer.get_feature_names_out()
    tfidf_scores = dict(zip(feature_names, tfidf_matrix.sum(axis=0).A1))
    top_words = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:20]

    plt.figure(figsize=(12, 6))
    plt.bar(*zip(*top_words))
    plt.xticks(rotation=45, ha='right')
    plt.title('Top Words by TF-IDF')
    plt.tight_layout()
    plt.show()

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(top_words))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Top TF-IDF Words')
    plt.show()

    return top_words

tfidf_words = tfidf_analysis(processed_texts)
```

# 4. Word2Vec analysis

```python
def word2vec_analysis(texts):
    model = Word2Vec(sentences=texts, vector_size=100, window=5, min_count=1, workers=4)

    # Get the most common word in the vocabulary
    most_common_word = model.wv.index_to_key[0]

    target_word = 'אבחנה'
    if target_word not in model.wv.key_to_index:
        print(f"\nWarning: '{target_word}' not found in vocabulary. Using '{most_common_word}' instead.")
        target_word = most_common_word

    try:
        similar_words = model.wv.most_similar(target_word, topn=10)

        print(f"\nWord2Vec Analysis Results:")
        print(f"Words most similar to '{target_word}':")
        print("-" * (25 + len(target_word)))
        for word, score in similar_words:
            print(f"{word:<15} {score:.4f}")
    except KeyError as e:
        print(f"\nError: {e}")
        print("Unable to perform similarity analysis. The vocabulary might be too small or the word might not exist.")
        similar_words = []

    # Print vocabulary statistics
    print(f"\nVocabulary size: {len(model.wv.key_to_index)}")
    print(f"Top 10 most frequent words: {', '.join(model.wv.index_to_key[:10])}")

    return similar_words

word2vec_words = word2vec_analysis(processed_texts)
```

# 5. Autoencoder for important words

```python
def autoencoder_important_words(texts, n_top_words=20, max_features=5000, encoding_dim=32, epochs=50, batch_size=32):
    # Ensure texts is a list of strings
    if isinstance(texts[0], list):
        texts = [' '.join(text) for text in texts]
    elif not all(isinstance(text, str) for text in texts):
        texts = [str(text) for text in texts]

    # TF-IDF Vectorization
    vectorizer = TfidfVectorizer(max_features=max_features)
    try:
        tfidf_matrix = vectorizer.fit_transform(texts).toarray()
    except ValueError as e:
        print(f"Error in TF-IDF vectorization: {e}")
        return []

    input_dim = tfidf_matrix.shape[1]

    # Build the autoencoder
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(encoding_dim, activation='relu')(input_layer)
    decoded = Dense(input_dim, activation='sigmoid')(encoded)
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')

    # Train the autoencoder
    try:
        history = autoencoder.fit(
            tfidf_matrix, tfidf_matrix,
            epochs=epochs,
            batch_size=batch_size,
            shuffle=True,
            validation_split=0.2,
            verbose=0
        )
    except tf.errors.ResourceExhaustedError:
        print("Error: Not enough memory. Try reducing batch_size or max_features.")
        return []

    # Extract important words
    decoder_weights = autoencoder.layers[-1].get_weights()[0]
    importance_scores = np.sum(np.abs(decoder_weights), axis=1)
    feature_names = vectorizer.get_feature_names_out()
    important_words = [(feature_names[i], importance_scores[i]) for i in importance_scores.argsort()[::-1][:n_top_words]]

    # Print results
    print("\nAutoencoder Analysis Results:")
    print(f"Top {n_top_words} important words:")
    print("-" * 30)
    for word, score in important_words:
        print(f"{word:<20} {score:.4f}")

    # Print training summary
    print(f"\nTraining summary:")
    print(f"Final loss: {history.history['loss'][-1]:.4f}")
    print(f"Final validation loss: {history.history['val_loss'][-1]:.4f}")

    return important_words

autoencoder_words = autoencoder_important_words(processed_texts)
```

# 6. Compare results of 3, 4, and 5

```python
def compare_word_importance_methods(tfidf_words, word2vec_words, autoencoder_words):
    tfidf_set = set(word for word, _ in tfidf_words)
    word2vec_set = set(word for word, _ in word2vec_words)
    autoencoder_set = set(word for word, _ in autoencoder_words)

    plt.figure(figsize=(10, 10))
    venn3([tfidf_set, word2vec_set, autoencoder_set],
          set_labels=('TF-IDF', 'Word2Vec', 'Autoencoder'))
    plt.title("Comparison of Important Words Across Methods")
    plt.show()

    print("Words common to all methods:", tfidf_set.intersection(word2vec_set, autoencoder_set))


compare_word_importance_methods(tfidf_words, word2vec_words, autoencoder_words)
```

# 7. Named Entity Recognition (NER)

```python

def perform_ner(texts, model_name="he_core_news_sm", fallback_model="en_core_web_sm"):
    try:
        nlp = spacy.load(model_name)
    except OSError as e:
        print(f"Error: Could not load the model '{model_name}'. {e}")
        print(f"Attempting to use fallback model '{fallback_model}'.")
        try:
            nlp = spacy.load(fallback_model)
        except OSError as e:
            print(f"Error: Could not load the fallback model '{fallback_model}'. {e}")
            print("Please install a spaCy model: python -m spacy download <model_name>")
            return []

    all_entities = []

    for i, text in enumerate(texts):
        if isinstance(text, list):
            text = ' '.join(text)
        try:
            doc = nlp(text)
            entities = [(ent.text, ent.label_) for ent in doc.ents]
            all_entities.extend(entities)
        except Exception as e:
            print(f"Error processing text {i}: {e}")

    entity_counts = Counter(label for _, label in all_entities)

    print("\nNamed Entity Recognition Results:")
    print("---------------------------------")
    print(f"Using model: {nlp.meta['lang']}_{nlp.meta['name']}")
    for label, count in entity_counts.most_common():
        print(f"{label:<10}: {count}")

    # Print some example entities
    print("\nExample entities:")
    for entity, label in set(all_entities[:10]):  # Show up to 10 unique entities
        print(f"{entity:<20} : {label}")

    return all_entities

ner_results = perform_ner(texts)
```

# 8. Exploratory Data Analysis (EDA)

```python

def perform_eda(data):
    print("Performing Exploratory Data Analysis...")

    if isinstance(data, pd.DataFrame):
        df = data
        if 'timestamp' not in df.columns or 'text' not in df.columns:
            raise ValueError("DataFrame must contain 'timestamp' and 'text' columns")
    elif isinstance(data, list):
        timestamps = pd.date_range(start='2023-01-01', periods=len(data), freq='H')
        df = pd.DataFrame({
            'timestamp': timestamps,
            'text': data
        })
    else:
        raise ValueError("Input must be either a DataFrame or a list of texts")

    plt.style.use('seaborn')

    # Daily counts
    daily_counts = df.groupby(df['timestamp'].dt.date).size()
    plt.figure(figsize=(12, 6))
    daily_counts.plot(kind='line')
    plt.title('Number of Transcripts per Day')
    plt.xlabel('Date')
    plt.ylabel('Number of Transcripts')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Hourly counts
    hourly_counts = df.groupby(df['timestamp'].dt.hour).size()
    plt.figure(figsize=(12, 6))
    hourly_counts.plot(kind='bar')
    plt.title('Number of Transcripts by Hour of Day')
    plt.xlabel('Hour')
    plt.ylabel('Number of Transcripts')
    plt.xticks(range(0, 24))
    plt.tight_layout()
    plt.show()

    # Weekday counts
    weekday_counts = df.groupby(df['timestamp'].dt.dayofweek).size()
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

    # Ensure weekday_counts has 7 elements
    full_week_counts = pd.Series(0, index=range(7))
    full_week_counts.update(weekday_counts)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=days, y=full_week_counts.values)
    plt.title('Number of Transcripts by Day of Week')
    plt.xlabel('Day of Week')
    plt.ylabel('Number of Transcripts')
    plt.tight_layout()
    plt.show()

    # Summary Statistics
    print("\nSummary Statistics:")
    print(f"Total number of transcripts: {len(df)}")
    print(f"Date range: from {df['timestamp'].min()} to {df['timestamp'].max()}")
    print(f"\nBusiest day: {daily_counts.idxmax()} with {daily_counts.max()} transcripts")
    print(f"Busiest hour: {hourly_counts.idxmax()} with {hourly_counts.max()} transcripts")
    print(f"Busiest day of week: {days[full_week_counts.idxmax()]} with {full_week_counts.max()} transcripts")


df = pd.DataFrame({
      'timestamp': pd.date_range(start='2023-01-01', periods=100, freq='H'),
      'text': texts
    })
perform_eda(df)
```

# 9. Keyword search and database structuring

```python

def keyword_search_and_structure(texts: List[str], keywords: Dict[str, List[str]], fields: List[str]) -> pd.DataFrame:
    """
    Search for keywords in texts and structure the data into a DataFrame.

    Args:
    texts (List[str]): List of text documents to search through.
    keywords (Dict[str, List[str]]): Dictionary mapping fields to lists of keywords.
    fields (List[str]): List of fields to include in the output.

    Returns:
    pd.DataFrame: Structured data with fields as columns.
    """
    print(f"Starting keyword search and structuring for {len(texts)} texts...")
    structured_data = []
    for i, text in enumerate(texts):
        print(f"Processing text {i+1}/{len(texts)}...")
        record = {field: None for field in fields}  # Initialize all fields to None
        for field, field_keywords in keywords.items():
            for keyword in field_keywords:
                pattern = rf"{re.escape(keyword)}:?\s*(.*?)(?=\n|$)"
                match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)
                if match:
                    record[field] = match.group(1).strip()
                    print(f"  Found match for field '{field}': {record[field]}")
                    break  # Stop searching for this field once a match is found
            if record[field] is None:
                print(f"  No match found for field '{field}'")
        structured_data.append(record)

    df = pd.DataFrame(structured_data)
    print(f"Keyword search completed. Created DataFrame with shape {df.shape}")
    return df

def create_database(df: pd.DataFrame, db_name: str = 'medical_summaries.db') -> None:
    """
    Create an SQLite database from a DataFrame.

    Args:
    df (pd.DataFrame): DataFrame to be stored in the database.
    db_name (str): Name of the database file to be created.

    Returns:
    None
    """
    try:
        conn = sqlite3.connect(db_name)
        df.to_sql('summaries', conn, if_exists='replace', index=False)
        conn.close()
        print(f"Database created successfully: {db_name}")
    except sqlite3.Error as e:
        print(f"An error occurred while creating the database: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")



keywords = {
        'patient_name': ['שם החולה', 'שם המטופל'],
        'age': ['גיל'],
        'diagnosis': ['אבחנה', 'דיאגנוזה'],
        'medications': ['תרופות', 'טיפול תרופתי'],
        'symptoms': ['תסמינים', 'סימפטומים'],
        'lab_results': ['תוצאות מעבדה', 'בדיקות מעבדה']
      }
fields = list(keywords.keys())
structured_df = keyword_search_and_structure(texts, keywords, fields)
create_database(structured_df)
```

# 10. Text Summarization

```python
import torch
from transformers import pipeline
from tqdm import tqdm

def summarize_texts(texts, max_length=14, min_length=0, batch_size=8):
    # Check if CUDA is available
    device = 0 if torch.cuda.is_available() else -1
    print(f"Using device: {'CUDA' if device == 0 else 'CPU'}")

    # Initialize the translation pipeline
    print("Initializing translation pipeline...")
    translator = pipeline("translation", model="facebook/mbart-large-50-many-to-many-mmt", device=device)

    # Initialize the summarization pipeline
    print("Initializing summarization pipeline...")
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

    summaries = []
    error_count = 0

    # Process texts in batches with a progress bar
    total_batches = (len(texts) + batch_size - 1) // batch_size
    for i in tqdm(range(0, len(texts), batch_size), total=total_batches, desc="Processing batches"):
        batch = texts[i:i+batch_size]
        try:
            # Translate Hebrew to English
            translations = translator(batch, src_lang="he_IL", tgt_lang="en_XX", max_length=512, truncation=True)
            translated_texts = [t['translation_text'] for t in translations]

            # Summarize the English texts
            summaries_batch = summarizer(translated_texts,
                                         max_length=max_length,
                                         min_length=min_length,
                                         do_sample=False)
            summaries.extend([s['summary_text'] for s in summaries_batch])
        except Exception as e:
            print(f"\nError processing batch {i//batch_size + 1}: {str(e)}")
            summaries.extend(["Error in summarization"] * len(batch))
            error_count += 1

    print("\n" + "="*50)
    print(f"Text summarization completed.")
    print(f"Total texts processed: {len(texts)}")
    print(f"Summaries created: {len(summaries)}")
    print(f"Batches with errors: {error_count}")
    print("="*50 + "\n")

    print("Sample of summaries:")
    for i, summary in enumerate(summaries[:5], 1):
        print(f"{i}. {summary}")

    if len(summaries) > 5:
        print("...")
        print(f"{len(summaries)}. {summaries[-1]}")

    return summaries

# Usage
summaries = summarize_texts(texts)
```

# 11. GPT model for Question Answering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm


def gpt_question_answering(texts, questions):
    device = 0 if torch.cuda.is_available() else -1
    print(f"Using device: {'CUDA' if device == 0 else 'CPU'}")

    print("Initializing question-answering pipeline...")
    qa_pipeline = pipeline("question-answering", model="deepset/xlm-roberta-large-squad2", device=device)

    results = {}
    total_questions = len(questions)
    total_texts = len(texts)

    print(f"\nProcessing {total_questions} questions for {total_texts} texts...")

    for i, question in enumerate(questions, 1):
        print(f"\nQuestion {i}/{total_questions}: '{question}'")
        answers = []

        for j, text in enumerate(tqdm(texts, desc=f"Processing texts for Q{i}", leave=False)):
            result = qa_pipeline(question=question, context=text, device=device)
            answers.append(result['answer'])

            if j < 2 or j == len(texts) - 1:  # Print first two and last answer for each question
                print(f"  Text {j+1}: {result['answer'][:50]}{'...' if len(result['answer']) > 50 else ''}")

        results[question] = answers
        print(f"Completed question {i}: Found {len(answers)} answers")

    print("\n" + "="*50)
    print("Question Answering Summary:")
    print(f"Total questions processed: {total_questions}")
    print(f"Total texts analyzed: {total_texts}")
    print(f"Total answers generated: {sum(len(a) for a in results.values())}")
    print("="*50)

    return results

questions = [
    "מהי האבחנה העיקרית?",
    "אילו תרופות ניתנו?",
    "מהם התסמינים העיקריים?",
    "מהן תוצאות הבדיקות?"
]
qa_results = gpt_question_answering(texts, questions)

print("\nSample of results:")
for question, answers in list(qa_results.items())[:2]:  # Print results for first two questions
    print(f"\nQuestion: {question}")
    print(f"Sample answers (first 3):")
    for answer in answers[:3]:
        print(f"- {answer[:50]}{'...' if len(answer) > 50 else ''}")
    if len(answers) > 3:
        print("...")
```

# 12. Compare summarization and question answering results

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re

def is_hebrew(text):
    return any("\u0590" <= char <= "\u05FF" for char in text)

def compare_summary_qa(summaries, qa_results):
    print("Initializing TF-IDF vectorizer...")
    vectorizer = TfidfVectorizer()

    print("Calculating summary vectors...")
    summary_vectors = vectorizer.fit_transform(summaries)

    for question, answers in qa_results.items():
        print(f"\nQuestion: {question}")

        print("Calculating answer vectors...")
        answer_vectors = vectorizer.transform(answers)

        print("Computing similarities...")
        similarities = cosine_similarity(summary_vectors, answer_vectors)

        for i, (summary, similarity_scores) in enumerate(zip(summaries, similarities)):
            most_similar_answer = answers[similarity_scores.argmax()]

            print(f"\nSummary {i+1}:")
            if is_hebrew(summary):
                print(f"Summary: {summary}")
            else:
                print(f"Summary: {summary}")

            if is_hebrew(most_similar_answer):
                print(f"Most similar answer: {most_similar_answer}")
            else:
                print(f"Most similar answer: {most_similar_answer}")

            print(f"Similarity score: {similarity_scores.max():.2f}")

    print("\nComparison completed.")

# Usage
compare_summary_qa(summaries, qa_results)
```
